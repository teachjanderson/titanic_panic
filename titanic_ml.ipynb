```python
from pathlib import Path
import pandas as pd
```

```python
# Load csv
# training data

ml_training_data = Path('Resources/tra_in.csv')

df_training = pd.read_csv(ml_training_data)

df_training.head(10)
```

```python
# testing data

ml_testing_data = Path('Resources/te_st.csv')

df_testing = pd.read_csv(ml_testing_data)

df_testing.head(10)
```

# Prepare and clean data

```python
# check for null values in both training and testing datasets

# training data nulls
df_training.isnull().sum()
```

```python
# test data nulls
df_testing.isnull().sum()
```

```python
# Fix missing values
# Replacing the missing values in the “Age” column with the mean value

df_testing['Age'].fillna(df_testing['Age'].mean(), inplace=True)
df_training['Age'].fillna(df_training['Age'].mean(), inplace=True)

# Chose not to drop 'Cabin' despite its missing values
# find the mode value of 'Cabin'
print(df_testing['Cabin'].mode())
print(df_testing['Cabin'].mode())   
```

```python
# Replacing the missing values in the “Cabin” column with mode value
df_testing['Cabin'].fillna(df_testing['Cabin'].mode()[0], inplace=True)
df_training['Cabin'].fillna(df_training['Cabin'].mode()[0], inplace=True)
```

```python
# Replacing the missing values in the “Embarked” column with mode value
df_testing['Embarked'].fillna(df_testing['Embarked'].mode()[0], inplace=True)
df_training['Embarked'].fillna(df_training['Embarked'].mode()[0], inplace=True)
```

```python
# Replacing the missing values in the “Fare” column with the mean value
df_testing['Fare'].fillna(df_testing['Fare'].mean(), inplace=True)
df_training['Fare'].fillna(df_training['Fare'].mean(), inplace=True)
```

```python
# Confirm 
# test data nulls
df_testing.isnull().sum()
```
